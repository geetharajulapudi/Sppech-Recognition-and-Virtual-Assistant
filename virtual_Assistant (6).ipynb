```python
import pyttsx3 as p
import speech_recognition as sr
import randfacts
import pyjokes
import datetime
import requests
import webbrowser
import pywhatkit
import calendar
import time
import subprocess
from imdb import IMDb
from textblob import TextBlob


#searching in wikipedia
class infow:
    def __init__(self):
        self.edge_path = "C:\\Program Files (x86)\\Microsoft\\Edge\\Application\\msedge.exe"

    def search_wikipedia(self, query):
        url = f"https://en.wikipedia.org/wiki/{query.replace(' ', '_')}"
        subprocess.Popen([self.edge_path, url])
        time.sleep(15)


#joke
def joke():
    url = "https://official-joke-api.appspot.com/random_joke"
    json_data = requests.get(url).json()
    array =["" , ""]
    array[0] = json_data["setup"]
    array[1] = json_data["punchline"]
    return array

#opening google
def open_google():
    webbrowser.open("https://www.google.com")
    time.sleep(10)

#opening gmail
def open_gmail():
    webbrowser.open("https://mail.google.com")
    time.sleep(10)

#opening googlemaps
def open_maps():
    webbrowser.open("https://www.google.com/maps")
    time.sleep(10)

#write a note
def note(text=None):
    if text:
        date = datetime.datetime.now()
        file_name = str(date).replace(":", "-") + "-note.txt"
        with open(file_name, "w") as f:
            f.write(text)
        subprocess.Popen(["notepad.exe", file_name])
        time.sleep(20)
    else:
        subprocess.Popen(["notepad.exe"])

#movie review
def get_movie_reviews(movie_name):
    ia = IMDb()
    movie = ia.search_movie(movie_name)[0]
    ia.update(movie, 'reviews')
    reviews = movie.get('reviews', [])[:1]    
    webbrowser.open(f"https://www.imdb.com/title/tt{movie.movieID}/reviews")
    return reviews

#Displaying current year calender
def current_cal(text_storage, data, text):
    text_storage['content'] = data + '\n' + text
    print(text_storage['content'])


engine = p.init()
rate = engine.getProperty('rate') #sentence time rate
engine.setProperty('rate',180)
voices = engine.getProperty('voices')
engine.setProperty('voice',voices[1].id)
#print(voices)

def speak(text):
    engine.say(text)
    engine.runAndWait()

r = sr.Recognizer()
m = sr.Microphone()

hour = int(datetime.datetime.now().hour)
if hour >= 0 and hour <= 12:
    print("Hello! Good Morning, How are you ?")
    speak("Hello! Good Morning, How are you ?")
elif hour >= 12 and hour <= 16:
    print("Hello! Good Afternoon, How are you ?")
    speak("Hello! Good Afternoon, How are you ?")
else:
    print("Hello! Good Evening, How are you ?")
    speak("Hello! Good Evening, How are you ?")



try:
    print("listening..")
    with m as source: r.energy_threshold =10000
    with m as source: r.adjust_for_ambient_noise(source,1.2)
    with m as source: audio = r.listen(source)  
    text = r.recognize_google(audio)
    print(text)
except:
    pass
    
try:   
    analysis = TextBlob(text)
    if analysis.sentiment.polarity < 0:
        print("Negative")
    else:
        print("Positive")
except:
    pass

try:
    if "what" and "about" and "you" in text:
        print("I am also having a good day")
        speak("I am also having a good day")
except:
    pass
    

def voice_Assistant():
    try:
        print("What can I do for You ?")
        speak("What can I do for You ?")
        try:
            with m as source: r.energy_threshold =10000
            with m as source: r.adjust_for_ambient_noise(source,1.2)
            print("listening....")
            with m as source: audio = r.listen(source) 
            text1 = r.recognize_google(audio)
            text2 = text1.lower()
            print("You said" ,text2)
        except:
            pass

        if "information" in text2 or "who is" in text2 or "search" in text2 :
            print("You need information related to which topic?")
            speak("You need information related to which topic?")
            with m as source: audio = r.listen(source)
            info = r.recognize_google(audio)
            print("Searching {} in wikipedia".format(info))
            speak("Searching {} in wikipedia".format(info))
            assist = infow()
            assist.search_wikipedia(info)
            return
        
        elif "play" in text2 or "video" in text2 or "youtube" in text2 or "song" in text2:
            print("You want me to play which video?")
            speak("You want me to play which video?")
            with m as source: audio = r.listen(source)
            video = r.recognize_google(audio)
            song = video.replace('play','')
            pywhatkit.playonyt(song)
            print(f"playing {song} on youtube")
            speak(f"playing {song} on youtube")
            time.sleep(30)
            return
        
        elif "fact" in text2 or "facts" in text2 or "randomfact" in text2:
            print("sure ")
            speak("sure ")
            ran = randfacts.get_fact()
            print("Did you know that ?"+ran)
            speak("Did you know that ?" +ran)
            return      

        elif "joke" in text2 or "jokes" in text2 or "fun" in text2 or "laugh" in text2 :
            print("Get ready to laugh")
            speak("Get ready to laugh")
            array = joke()
            print(array[0])
            speak(array[0])
            print(array[1])
            speak(array[1])
            return
        
        elif "open Google" in text2 or "google" in text2:
            print("Opening Google..")
            speak("Opening Google..")
            open_google()
            return      

        elif "gmail" in text2 or "open Email" in text2 or "open Gmail" in text2 :
            print("Opening Mail..")
            speak("Opening Mail..")
            open_gmail()
            return
        
        elif "open Maps" in text2 or "location" in text2 or "maps" in text2:
            print("Opening Google Maps..")
            speak("Opening Google Maps..")
            open_maps()
            return
      
        elif "note" in text2 or "Notepad" in text2 or "to write a note" in text2 or "notepad" in text2:
            print("opening notepad..")
            speak("opening notepad..")
            note()
            note("This is a sample Note")
            return
    
        elif "time" in text2 or "current time" in text2:
            time =datetime.datetime.now().strftime("%I:%M%p")
            print("Current Time is " +time)
            speak("Current Time is " +time)
            return

        elif "date" in text2 or "what is today date" in text2:
            today_date = datetime.datetime.now()
            print("Today is " + today_date.strftime("%d") + " of " + today_date.strftime("%b") + " and Its Currently " + (today_date.strftime("%Y"))) 
            speak("Today is " + today_date.strftime("%d") + " of " + today_date.strftime("%b") + " and Its Currently " + (today_date.strftime("%Y")))
            return

        elif "calendar" in text2 or "current year calendar" in text2 :
            print("Displaying current year calender.")
            speak("Displaying current year calender.")
            text_storage = {'content': ''}
            current_cal(text_storage, calendar.calendar(2024), "")
            return

        elif "who are you " in text2 or "your name" in text2 or "name" in text2:
            print("I am your Voice Assistant.")
            speak("I am your Voice Assistant.")
            return
        
        elif "review" in text2 or "rating" in text2 or "moviereview" in text2 :
            print("You need which movie review ?")
            speak("You need which movie review ?")
            with m as source: audio = r.listen(source)
            movie_name = r.recognize_google(audio)
            print("Searching {} review".format(movie_name))
            speak("Searching {} in review".format(movie_name))
            mov = get_movie_reviews(movie_name)
            print(mov)
            speak(mov)
            return

        elif "good bye" in text2 or "bye" in text2 or "see you" in text2 or "later" in text2:
            print("Good Bye")
            speak("Good Bye")
            return "break"
            
        elif "Thankyou" in text2 or "thanks" in text2 or "thanks for giving response" in text2:
            print("Welcome!")
            speak("Welcome!")

        else:
            print("Sorry! I did not understand!. could you repeat again ?.")
            speak("Sorry! I did not understand!. could you repeat again ?.")
            return

    except Exception as exception:
        speak("Sorry! I did not quite catch that .")


while True:
    res = voice_Assistant()
    if res == "break":
        break

```
